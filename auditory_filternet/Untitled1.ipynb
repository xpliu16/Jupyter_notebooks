```python
import numpy as np
import matplotlib.pyplot as plt

a = 1
b = 1    # controls width
f = 20
n = 2    # controls shape
phi = np.pi/2
t = np.arange(0,1, 0.005)
gt = a * (t**(n-1)) * np.exp(-2*np.pi*b*t)* np.cos(2*np.pi*f*t+phi)
plt.plot(gt)
```

```python
t = np.arange(0, 3, 0.001)
f = 0.5
# f term in amplitude kind of normalizes its size, although you will normalize anyways later
gt2 = (f*t)**2*np.exp(-6*f*t) * np.sin(2*np.pi*f*t)
#gt2 -= np.mean(gt2)
plt.plot(gt2)
```

```python
# Unmodulated - of course it really is modulated
t = np.arange(0, 2, 0.001)
f = 0.5

gt2 = (f*t)**2*np.exp(-6.5*f*t)* np.sin(2*np.pi*f*t)
plt.plot(gt2)
```

```python
f = 0.5
Lambda = 1/f
sigma_t = Lambda/4
gt3 = np.exp(-.5 * ((t-0.5*Lambda)**2/sigma_t**2)) * np.cos(2*np.pi*f*(t-0.5*Lambda)+np.pi/2)
plt.plot(gt3)
```

```python
# for if temporal modulation is 0
t = np.arange(0, 2, 0.001)
f = 2.5

gt4 = (f*t)**2*np.exp(-10*f*t)* np.sin(2*np.pi*f*t)
plt.plot(t, gt4)
np.min(gt4)
```

```python
from scipy.io.wavfile import write
sr = 32000

t = np.arange(0,1.9,1/sr)
f = 50 + (6000-50)*t/1.9
stim = 1* np.cos(2*np.pi*f*t)

ramp_t = np.arange(0,0.010,1/sr)
ramp = (1-np.cos(2*np.pi*50*ramp_t))/2
plt.plot(ramp_t, ramp)
stim[0:0+len(ramp)] *= ramp  
offramp = np.flip(ramp)
stim[-len(offramp)-1:-1] *= offramp

scaled = np.int16(stim / np.max(np.abs(stim)) * 32767)
plt.plot(scaled)

scaled = np.flip(scaled)

plt.show()
#scaled = np.int16(sent1 * 32767)
write('/home/ping/Filternet_improvement/FM_sweep_stim.wav', sr, scaled)
```
