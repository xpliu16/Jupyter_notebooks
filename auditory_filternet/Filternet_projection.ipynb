```python
import numpy as np
from matplotlib import pyplot as plt

#t_mod_set = np.geomspace(1, 20.1587368, 14)
t_mod_set = np.geomspace(2, 16, 10)
#t_mod_set = np.concatenate((t_mod_set, t_mod_set))     # For up versus down orientations
t_mod_set = np.insert(t_mod_set, 0, 0)
print(t_mod_set)
print(t_mod_set.shape)
```

```python
sp_mod_set = np.geomspace(0.25, 2, 13)
#sp_mod_set = np.concatenate((sp_mod_set, sp_mod_set)) 
#sp_mod_set = np.geomspace(0.25, 2, 4)
sp_mod_set = np.insert(sp_mod_set, 0, 0)
sp_mod_set = np.flip(sp_mod_set)     # Make it look more like spectrotemporal modulation space
print(sp_mod_set)
print(sp_mod_set.shape)
```

```python
n_filts = len(t_mod_set)*len(sp_mod_set) + (len(t_mod_set)-1)*(len(sp_mod_set)-1) 
print('Number of distinct filters (before considering center frequency):', n_filts)

# Are there fewer nodes for filters with smaller f range that is permitted? 
# But then vertical filters would be underrepresented - opt to keep same number for each filter type 
# Plus, the sampling stage is when we think about representation

factor = 20000//n_filts
n_nodes = factor * n_filts
print('Number of nodes:', n_nodes)
print('Multiplicative factor:', factor)
```

```python
t_mod_list, sp_mod_list = [v.flatten() for v in np.meshgrid(t_mod_set, sp_mod_set)]

#t_mod_list2, sp_mod_list2 = [v.flatten() for v in np.meshgrid(t_mod_set[1:], sp_mod_set[:-1])]
#t_mod_list = np.concatenate((t_mod_list, t_mod_list2))
#sp_mod_list = np.concatenate((sp_mod_list, sp_mod_list2))

print(t_mod_list.shape)
print(sp_mod_list.shape)
t_mod_list
```

```python
from bmtk.simulator.filternet.lgnmodel.waveletfilter import WaveletFilter

# See where spectral Gaussian sigma_f drops to threshold_rel of peak
# It does depend on interaction with spectral modulation... really should be calculated empirically
framerate = 0.001

min_freq = 50
max_freq = 8000
n_freqs = 131

row_range = np.log2(np.geomspace(min_freq, max_freq, n_freqs)/min_freq)
col_range = np.arange(0, 2, framerate)
sigma_f = 1/sp_mod_list/4
theta = np.arctan(sp_mod_list / t_mod_list)
Lambda = 1/np.linalg.norm([t_mod_list, sp_mod_list], axis=0)

# Assumes filters are set in the middle of the log freq range and therefore unlikely to be truncated on either side.
translate = (row_range[-1]-row_range[0])/2
print("translate:", translate)
delta_freq = row_range[1]-row_range[0]
print("delta_freq:", delta_freq)
min_freq_filt = np.empty_like(sp_mod_list)
max_freq_filt = np.empty_like(sp_mod_list)

#sp_mod_list_all = np.repeat(sp_mod_list, factor)
#print(sp_mod_list_all.shape)
#print(sp_mod_list_all[1:50])
#t_mod_list_all = np.repeat(t_mod_list, factor)
sp_mod_list_all = np.array([])
t_mod_list_all = np.array([])
directions_all = np.array([])
freqs_all = np.array([])

#fig, ax = plt.subplots(len(t_mod_set), len(sp_mod_set), sharey=True, sharex=True, figsize=(12,8))

fig = plt.figure(figsize=(16,12))
ax = [None]*len(t_mod_list)

for i in range(len(t_mod_set)*len(sp_mod_set)):
    ax[i] = fig.add_subplot(len(sp_mod_set),len(t_mod_set), i+1)
    if sp_mod_list[i] == 0:
        # Temporal only filter - width is Inf
        # Set spectral width equal to that of filter two above it (0.25 spectral mod)
        sigma_f[i] = 1/sp_mod_set[-2]/4
        print("temporal filter:", sigma_f[i])
    if t_mod_list[i] == 0:
        # Spectral only filter
        psi = 0
    else: 
        psi = -np.pi/2
    # Make filter for t_mod_list[i] and sp_mod_list[i]
    f = WaveletFilter(translate=translate, sigma_f=sigma_f[i], b_t=6.5, order_t=3, theta=theta[i], 
                      Lambda=Lambda[i], psi=psi, delay=0.0,
                 amplitude= 1., direction=1)
    #ax_temp = ax[i//len(t_mod_set), i%len(t_mod_set)]
    #f.get_kernel(row_range, col_range, threshold_rel=0.05, show_plot=True, plot_ax = ax[i//len(t_mod_set), i%len(t_mod_set)])
    kernel = f.get_kernel(row_range, col_range, threshold_rel=0.05, show_plot=True, plot_ax = ax[i])
    #f.get_kernel(row_range, col_range, threshold_rel=0.05, show_plot=True)
    #ax[i].set_title('f_sm={}, f_tm={:.2g}'.format(sp_mod_list[i], t_mod_list[i]))
    
    ax[i].set_xlim(0,0.5)
    ax[i].set(xlabel=None)
    
    # Based on Gaussian fall from max to 0.05 of peak, doesn't take into account oscillation and phase, but pretty close
    extent_calc = 2*np.sqrt(-2*np.log(0.05)*sigma_f[i]**2)
    print("Extent_calc:", extent_calc)
    extent_ind = len(np.unique(kernel.row_inds))
    extent_freq = extent_ind*delta_freq     # spectral width of thresholded filter (in log freq units)
    print("Extent freq:", extent_freq)
    min_freq_filt[i] = row_range[0]+extent_freq/2
    max_freq_filt[i] = row_range[-1]-extent_freq/2
    #print('sp_mod:', sp_mod_list[i])
    #print('t_mod:', t_mod_list[i])
    #print('sigma_f:', sigma_f[i])
    #print('extent_ind:', extent_ind)
    #print('extent_freq:', extent_freq)
    #print('min_freq:', min_freq_filt[i])
    #print('max_freq:', max_freq_filt[i])
    sp_mod_list_all = np.concatenate((sp_mod_list_all, np.repeat(sp_mod_list[i],factor)))
    t_mod_list_all = np.concatenate((t_mod_list_all, np.repeat(t_mod_list[i],factor)))
    freqs_all = np.concatenate((freqs_all, np.random.uniform(min_freq_filt[i], max_freq_filt[i], factor)))
    if (sp_mod_list[i]!=0) and (t_mod_list[i]!=0):
        directions_all = np.concatenate((directions_all, np.repeat([1],factor)))
        sp_mod_list_all = np.concatenate((sp_mod_list_all, np.repeat(sp_mod_list[i],factor)))
        t_mod_list_all = np.concatenate((t_mod_list_all, np.repeat(t_mod_list[i],factor)))
        directions_all = np.concatenate((directions_all, np.repeat([-1],factor)))
        freqs_all = np.concatenate((freqs_all, np.random.uniform(min_freq_filt[i], max_freq_filt[i], factor)))
    else:
        directions_all = np.concatenate((directions_all, np.repeat([0],factor)))
        

    #plt.gca().set_xlim(0, 1)
    #xlabel='Time(s)', ylabel='log(freq)'
    # duplicate each row and assign center freqs
    # Test a small batch scale up
    
#fig.tight_layout()
```
